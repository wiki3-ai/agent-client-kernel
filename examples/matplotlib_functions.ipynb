# Drawing Functions with Matplotlib

This notebook demonstrates how to use matplotlib to visualize various mathematical functions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Linear Function

Drawing a simple linear function: $y = 2x + 1$

```python
# Create x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y = 2 * x + 1

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x + 1', linewidth=2)
plt.title('Linear Function: y = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 2. Quadratic Function

Drawing a parabola: $y = x^2$

```python
# Create x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y = x ** 2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x²', linewidth=2)
plt.title('Quadratic Function: y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 3. Trigonometric Functions

Drawing sine and cosine functions

```python
# Create x values (from -2π to 2π)
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)

# Calculate y values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='sin(x)', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linewidth=2)
plt.title('Trigonometric Functions')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 4. Exponential Function

Drawing the exponential function: $y = e^x$

```python
# Create x values
x = np.linspace(-3, 3, 100)

# Calculate y values
y = np.exp(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = eˣ', linewidth=2)
plt.title('Exponential Function: y = eˣ')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

## 5. Multiple Functions on One Plot

Comparing different functions in the same graph

```python
# Create x values
x = np.linspace(-5, 5, 100)

# Calculate y values for multiple functions
y1 = x
y2 = x ** 2
y3 = x ** 3

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = x', linewidth=2)
plt.plot(x, y2, label='y = x²', linewidth=2)
plt.plot(x, y3, label='y = x³', linewidth=2)
plt.title('Comparing Polynomial Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.ylim(-10, 10)
plt.show()
```

## 6. Polar Plot

Drawing a function in polar coordinates: $r = \sin(2\theta)$

```python
# Create theta values
theta = np.linspace(0, 2 * np.pi, 300)

# Calculate r values
r = np.sin(2 * theta)

# Create the polar plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.plot(theta, r, linewidth=2)
ax.set_title('Polar Function: r = sin(2θ)', va='bottom')
plt.show()
```

## 7. Custom Styling

Creating a plot with custom colors, markers, and styling

```python
# Create x values
x = np.linspace(0, 10, 50)

# Calculate y values
y = np.exp(-0.3 * x) * np.sin(x)

# Create the plot with custom styling
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', markersize=8, linewidth=2,
         color='#2ecc71', label='Damped Sine Wave')
plt.title('Custom Styled Plot: Damped Sine Wave', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.show()
```
