```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
%matplotlib inline

# Golden ratio constant
GOLDEN_RATIO = (1 + np.sqrt(5)) / 2
```

```python
def draw_golden_rectangle(ax, x, y, width, height, depth, max_depth):
    """
    Draw a golden rectangle and recurse to create the fractal.
    
    Parameters:
    - ax: matplotlib axes
    - x, y: bottom-left corner coordinates
    - width, height: dimensions of the rectangle
    - depth: current recursion depth
    - max_depth: maximum recursion levels
    """
    # Draw the rectangle
    if depth == 0:
        # Outermost rectangle - draw in a distinct color
        rect = plt.Rectangle((x, y), width, height, fill=False, edgecolor='blue', linewidth=2)
        ax.add_patch(rect)
    else:
        # Inner rectangles
        rect = plt.Rectangle((x, y), width, height, fill=False, edgecolor='red', linewidth=1)
        ax.add_patch(rect)
    
    # If we haven't reached max depth, draw the next square/rectangle
    if depth < max_depth:
        # For golden rectangle: remove a square of size min(width, height)
        if width > height:
            # Width is longer, remove a square of size height
            new_width = width - height
            # The remaining rectangle has dimensions: height x (width - height)
            # Position it appropriately
            draw_golden_rectangle(ax, x + height, y, new_width, height, depth + 1, max_depth)
        else:
            # Height is longer or equal, remove a square of size width
            new_height = height - width
            # The remaining rectangle has dimensions: (height - width) x width
            draw_golden_rectangle(ax, x, y + width, width, new_height, depth + 1, max_depth)

def draw_golden_fractal(num_levels):
    """
    Draw the golden ratio fractal with specified number of recursion levels.
    
    Parameters:
    - num_levels: Number of recursion levels (1-10 recommended)
    """
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Initial golden rectangle (width:height = GOLDEN_RATIO:1)
    initial_width = 8
    initial_height = initial_width / GOLDEN_RATIO
    
    # Center the rectangle
    x_start = (10 - initial_width) / 2
    y_start = (8 - initial_height) / 2
    
    # Clear and set up the plot
    ax.clear()
    draw_golden_rectangle(ax, x_start, y_start, initial_width, initial_height, 0, num_levels)
    
    # Set plot properties
    ax.set_aspect('equal', 'box')
    ax.set_xlim(-0.5, 10)
    ax.set_ylim(-0.5, 8)
    ax.axis('off')
    
    title = f'Golden Ratio Fractal - {num_levels} Recursion Level(s)'
    ax.set_title(title, fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
```

JPW: This is a nice version but had `min=`1 which I changed to `min=0` and so now it is an even better demo.

```python
# Create interactive slider for recursion levels
interact(draw_golden_fractal, 
         num_levels=IntSlider(min=0, max=10, value=5, 
                             description='Levels:', 
                             style={'description_width': '100px'},
                             continuous_update=False))
```
